# Question 1. Write a code to find the province with the most hydropowers and plot them in a map

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import contextily as ctx

# Load CSV data
df = pd.read_csv("/content/sample_data/lab4data/hydropower-projects.csv")

# Count plants per province
province_counts = df['Province'].value_counts()
most_hydro_province = province_counts.idxmax()

# Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df["Longitude"], df["Latitude"]),
    crs="EPSG:4326"
)

# Load shapefile of Nepal's provinces
shapefile_path = "/content/sample_data/lab4data/province.shp"
gdf_provinces = gpd.read_file(shapefile_path).to_crs("EPSG:4326")

# Create plot
fig, ax = plt.subplots(figsize=(12, 10))

# Plot province boundaries
gdf_provinces.boundary.plot(ax=ax, edgecolor='black', linewidth=1)

# Plot all hydropower plants
gdf.plot(ax=ax, color='blue', markersize=30, alpha=0.6, label='Hydropower Plants')

# Highlight province with most hydropowers
highlight = gdf[gdf['Province'] == most_hydro_province]
highlight.plot(ax=ax, color='red', markersize=40, label=f'Most: {most_hydro_province}')

# Add basemap
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron, crs=gdf.crs)

# Decorations
# Axes and labels
ax.set_title("Hydropower Plants in Nepal", fontsize=16)
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(loc='lower right')
plt.figtext(0.95, 0.11, " Mohit Guragain", ha="right", fontsize=10, color='Black')
plt.tight_layout()
plt.show()
```

# **Write a code to scrape the following urls**
https://en.wikipedia.org/wiki/Markov_decision_process
https://en.wikipedia.org/wiki/Reinforcement_learning
https://en.wikipedia.org/wiki/Multi-agent_reinforcement_learning
https://en.wikipedia.org/wiki/Andrew_Barto
https://en.wikipedia.org/wiki/Richard_S._Sutton
https://pt.wikipedia.org/wiki/Aprendizagem_por_refor%C3%A7o

```python
import requests
from bs4 import BeautifulSoup
import re
from collections import Counter
import matplotlib.pyplot as plt
import textwrap



def scrape_wikipedia_text(url):
    """Scrape all visible paragraph texts from a Wikipedia URL and return concatenated text."""
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        content_div = soup.find('div', id='mw-content-text')

        paragraphs = content_div.find_all('p')
        texts = [p.get_text().strip() for p in paragraphs if p.get_text().strip()]
        full_text = ' '.join(texts)
        return full_text
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return ""

def preprocess_text(text):
    """Lowercase, remove non-alpha chars, and split into words."""
    text = text.lower()
    # Replace anything that's NOT a-z or apostrophe or whitespace with space
    text = re.sub(r"[^a-záéíóúâêîôûàèìòùãõç']+", ' ', text)
    words = text.split()
    return words

def plot_top_words(word_counts, top_n=20):
    """Plot the top N words by count."""
    common_words = word_counts.most_common(top_n)
    words, counts = zip(*common_words)

    plt.figure(figsize=(12, 6))
    plt.bar(words, counts, color='skyblue')
    plt.xticks(rotation=45, ha='right')
    plt.title(f"Top {top_n} Words by Frequency")
    plt.xlabel("Words")
    plt.ylabel("Count")
    plt.figtext(0.99, 0.01, "Mohit Guragain", ha='right', fontsize=12)
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    urls = [
        "https://en.wikipedia.org/wiki/Markov_decision_process",
        "https://en.wikipedia.org/wiki/Reinforcement_learning",
        "https://en.wikipedia.org/wiki/Multi-agent_reinforcement_learning",
        "https://en.wikipedia.org/wiki/Andrew_Barto",
        "https://en.wikipedia.org/wiki/Richard_S._Sutton",
        "https://pt.wikipedia.org/wiki/Aprendizagem_por_refor%C3%A7o"
    ]

    all_text = ""
    for url in urls:
        print(f"\n\n--- Scraping: {url} ---")
        text = scrape_wikipedia_text(url)

        # Print first 500 characters of scraped text to avoid flooding console
        print(textwrap.fill(text[:130]) + ('...' if len(text) > 500 else ''))

        all_text += text + " "

    words = preprocess_text(all_text)
    word_counts = Counter(words)

    plot_top_words(word_counts, top_n=20)
```

# **Question 3 Write a program to implement text clustering.**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import textwrap

# Step 1: Load the text file from Google Drive
file_path = '/content/sample_data/lab4data/question_3_data.txt'

with open(file_path, 'r', encoding='utf-8') as file:
    documents = file.readlines()

# Clean documents by stripping whitespace and removing empty lines
documents = [doc.strip() for doc in documents if doc.strip()]

# Step 2: Convert documents to TF-IDF vectors
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

# Step 3: Apply KMeans clustering
num_clusters = 3  # adjust as needed
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Step 4: Print clustering results with wrapped text
print("\n===== Text Clustering Results =====")
wrapper = textwrap.TextWrapper(width=120)

for i, doc in enumerate(documents):
    cluster = kmeans.labels_[i]
    wrapped_doc = wrapper.fill(doc)
    print(f"\n--- Document {i+1} (Cluster {cluster}) ---")
    print(wrapped_doc)

print("\n Mohit Guragain")
```

# **Question 4**
Download the data from MSNBC.
Write a program to count the frequency of each page category and plot a bar chart.
Identify and count all page-to-page transitions (e.g., FrontPage → News).
Plot a Sankey diagram to visualize the top 30 most common transitions.


```python
import matplotlib.pyplot as plt
from collections import Counter
from itertools import tee
import plotly.graph_objects as go

# File path
file_path = '/content/sample_data/lab4data/description.txt'

# Step 1: Load and clean data
with open(file_path, 'r') as file:
    sessions = [line.strip().split() for line in file if line.strip()]

# Step 2: Count frequency of each page category
page_counts = Counter()
for session in sessions:
    page_counts.update(session)

# Plot bar chart
plt.figure(figsize=(12, 6))
plt.bar(page_counts.keys(), page_counts.values(), color='skyblue')
plt.xlabel('Page Category')
plt.ylabel('Frequency')
plt.title('Page Category Frequency (MSNBC)')
plt.figtext(0.99, 0.01, "Mohit", ha="right", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Step 3: Count page-to-page transitions
transition_counts = Counter()
for session in sessions:
    for i in range(len(session) - 1):
        transition = (session[i], session[i + 1])
        transition_counts[transition] += 1

# Step 4: Sankey Diagram of Top 30 Transitions
top_transitions = transition_counts.most_common(30)

labels = list(set([src for src, _ in top_transitions] + [dst for _, dst in top_transitions]))
label_to_index = {label: i for i, label in enumerate(labels)}

source = [label_to_index[src] for src, dst in top_transitions]
target = [label_to_index[dst] for src, dst in top_transitions]
values = [count for (_, _), count in top_transitions]

fig = go.Figure(data=[go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color='black', width=0.5),
        label=labels
    ),
    link=dict(
        source=source,
        target=target,
        value=values
    ))])

fig.update_layout(title_text='Top 30 Page-to-Page Transitions (MSNBC)', font_size=12, annotations=[dict(text="Mohit", x=1, y=-0.1, showarrow=False, xanchor='right', yanchor='bottom', font=dict(size=14))]
)
fig.show()
```
